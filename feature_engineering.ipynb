```python
df = pd.read_csv("customers_with_orders.csv")
df.drop("customer_id", axis=1)
df = df[(df["order_status"] != "unavailable") & (df["order_status"] != "canceled")]
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
```

```python
split_date = df["order_purchase_timestamp"].quantile(0.8)
latest_date = df["order_purchase_timestamp"].max()
```

```python
df["estimated_delivery_days"] = (pd.to_datetime(df["order_estimated_delivery_date"]) - df["order_purchase_timestamp"]).dt.days
```

```python
df_train = df[df["order_purchase_timestamp"] < split_date]
df_test = df[df["order_purchase_timestamp"] >= split_date]
df_train.shape, df_test.shape
```

```python
df_train = df_train.groupby("customer_unique_id")
df_train = df_train.agg({"order_purchase_timestamp": ["count", "max"], "customer_city": "first", "customer_city": "first", "customer_zip_code_prefix": "first", "estimated_delivery_days": ["mean", "min", "max"]})
df_train["order_purchase_timestamp"]["count"] = df_train["order_purchase_timestamp"]["count"].astype(int)
df_train["date_diff"] = (latest_date - df_train["order_purchase_timestamp"]["max"]).dt.days
df_train.head()
```

```python
df_train.columns = ['_'.join(col).strip() for col in df_train.columns.values]
```

```python
x = pd.DataFrame()

x["customer_unique_id"] = df["customer_unique_id"]
x["customer_city"] = df["customer_city"]
x["customer_zip_code_prefix"] = df["customer_zip_code_prefix"]


x = x.join(df_train, on="customer_unique_id", how="left")
x = x.drop(["order_purchase_timestamp_max", "customer_city_first", "customer_zip_code_prefix_first"], axis=1)
x["order_purchase_timestamp_count"] = x["order_purchase_timestamp_count"].fillna(0).astype(int)
x["date_diff_"] = x["date_diff_"].fillna(x['date_diff_'].max() * 2).astype(int)
```

```python
x["estimated_delivery_days_max"] = x["estimated_delivery_days_max"].fillna(x['estimated_delivery_days_max'].max() * 2).astype(int)
x["estimated_delivery_days_mean"] = x["estimated_delivery_days_mean"].fillna(x['estimated_delivery_days_mean'].max() * 2).astype(int)
x["estimated_delivery_days_min"] = x["estimated_delivery_days_min"].fillna(x['estimated_delivery_days_min'].max() * 2).astype(int)
```

```python
df_y = pd.DataFrame()
df_y["customer_unique_id"] = df["customer_unique_id"]
```

```python
df_test = df_test.groupby("customer_unique_id")
df_test = df_test.agg({"order_purchase_timestamp": ["count"]})
df_test.columns = ['_'.join(col).strip() for col in df_test.columns.values]
df_y = df_y.join(df_test, on="customer_unique_id", how="left")
df_y["order_purchase_timestamp_count"] = df_y["order_purchase_timestamp_count"].fillna(0).astype(int)
df_y["target"] = df_y["order_purchase_timestamp_count"].apply(lambda x: 1 if x > 0 else 0)
```

```python
df_y = df_y.set_index("customer_unique_id").reindex(x["customer_unique_id"]).reset_index()
```

```python
x = x.drop(["customer_unique_id"], axis=1)
y = df_y["target"]
print(x.shape, y.shape)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

x_train_categorical = x_train[["customer_city"]]
enc = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
enc.fit(x_train_categorical)

# Transform and reset indices for proper alignment
x_train_categorical = pd.DataFrame(enc.transform(x_train_categorical), index=x_train.index)
x_test_categorical = pd.DataFrame(enc.transform(x_test[["customer_city"]]), index=x_test.index)

# Drop the original categorical column
x_train = x_train.drop(["customer_city"], axis=1)
x_test = x_test.drop(["customer_city"], axis=1)

# Concatenate the one-hot encoded data with the original DataFrames
x_train = pd.concat([x_train, x_train_categorical], axis=1)
x_test = pd.concat([x_test, x_test_categorical], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Ensure all column names in x_train and x_test are strings
x_train.columns = x_train.columns.astype(str)

# reg = LogisticRegression(max_iter=1000, random_state=42)
# reg_scores = cross_val_score(reg, x_train, y_train, cv=5, scoring="f1")
# print("Logistic Regression F1 Score: ", reg_scores.mean())

rf = RandomForestClassifier(random_state=42)
# rf_scores = cross_val_score(rf, x_train, y_train, cv=5, scoring="f1")
# print("Random Forest F1 Score: ", rf_scores.mean())
```

```python
rf.fit(x_train, y_train)
```

```python

# Ensure all column names in x_test are strings
x_test.columns = x_test.columns.astype(str)

y_pred = rf.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf.classes_)

disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")

clf_report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(clf_report).transpose()
report_df
```

```python
df_features = pd.DataFrame(rf.feature_importances_, index=x_train.columns, columns=["importance"])
df_features = df_features.sort_values("importance", ascending=False)

df_features.head(10).plot(kind="barh", figsize=(10, 6), color="skyblue")
```
